**Колебательный процесс**

```python
print(R)
C3 = 0.05 * C
Rcr = 2 * np.sqrt(L/C3)
print(Rcr)
T = 1.16 * 10 ** (-3)
im = 17.61 * 10 ** (-3)
im_t = 4.36 * 10 ** (-3)
print("1. T = ", T)
f = 1 / T
print("2. f = ", f)
w = 2 * np.pi * f
print("3. w = ", w)
delta = im / im_t
print("4. delta = im / im_t = ", delta)
thetha = np.log(delta)
print("5. theta =  ", thetha)
sigma = thetha / T
print("6. sigma = ", sigma)
omega = np.sqrt(w ** 2 + sigma ** 2)
print("7. omega = ", omega)
L_new = 1 / (omega ** 2 * C3)
print("8. L = ", L_new)
R_active = 2 * sigma * L_new
print("9. R = ", R_active)
```
